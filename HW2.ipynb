```python
import pandas as pd
import matplotlib.pyplot as plt
import wbgapi as wb
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
series_id = ["SE.ADT.LITR.FE.ZS", "SL.TLF.TOTL.FE.ZS", "SI.POV.NAHC", "SH.XPD.CHEX.PC.CD", "NY.GDP.PCAP.CD", "SH.DYN.MORT"]
country = "TUR"
```

```python
lit = wb.data.DataFrame(series_id[0], country)
lab = wb.data.DataFrame(series_id[1], country)
pov = wb.data.DataFrame(series_id[2], country)
health = wb.data.DataFrame(series_id[3], country)
gdp = wb.data.DataFrame(series_id[4], country)
mor = wb.data.DataFrame(series_id[5], country)
```

```python
lit.reset_index(drop=True, inplace = True)
lab.reset_index(drop= True, inplace = True)
pov.reset_index(drop=True,inplace = True)
health.reset_index(drop=True,inplace = True)
gdp.reset_index(drop=True,inplace = True)
mor.reset_index(drop=True, inplace= True)

master_df = pd.concat([lit,lab,pov,health,gdp,mor],axis=0)
master_df
```

```python
master_df.index = ["Literacy", "Labor",
                   "Poverty", "Health", "GDP", "Mortality"]
master_df
```

```python
master_df.isnull().sum()
master_df.describe()
```

```python
df = master_df.T
df.dropna(inplace=True)
df
```

```python
corr_matrix = df.corr()
corr_matrix
```

```python
x = df[["Labor",'Literacy', "Poverty","GDP", "Health"]]
y = df['Mortality']

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()
predictions = model.predict(x) 

model.summary()
```

```python
model1 = ols('Mortality ~ Labor + Literacy + Poverty + Health + GDP ', data=df).fit()
print(model1.summary())
sm.stats.anova_lm(model1)
```

```python
model1 = ols('Mortality ~ Poverty * Labor * Literacy + Health ', data=df).fit()
print(model1.summary())
sm.stats.anova_lm(model1)
```

### By looking at the correlation coefficients , R-squared score and the anova table, the model that gave the most accurate results among the models I tried with different rankings and variables is model1.

```python
model2 = ols('Mortality ~  Labor*Poverty  ', data=df).fit()
print(model2.summary())
sm.stats.anova_lm(model2)
```

```python
model3 = ols('Mortality ~ Labor * Literacy ', data = df).fit()
print(model3.summary())
sm.stats.anova_lm(model3)
```

```python
def open_close_price(ticker_symbol):
    temp_ticker = yf.Ticker(ticker_symbol)
    df_ticker = temp_ticker.history(period = 'max')
    col_names = df_ticker.columns
    bin_comp = []
    for i in range(len(df_ticker.index)):
        if df_ticker['Open'][i] < df_ticker['Close'][i]:
            bin_comp.append(1)
        else:
            bin_comp.append(0) 

    df_ticker['Day Positive'] = bin_comp
    df_ticker.drop(columns=col_names, inplace = True)
    return df_ticker
```

```python
gold = open_close_price('GC=F')
gold
```

```python
silver = open_close_price('SI=F')
silver
```

```python
credit = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data', header=None)
fn = {'+': 1, '-': 0}

X = credit.replace('?',0).iloc[:,[1,2,7,10,14]]
y = credit.iloc[:,15].map(lambda x: fn.get(x,0))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
logreg1 = LogisticRegression(max_iter = 200).fit(X_train,y_train)
y_pred1 = logreg1.predict(X_test)
print("Accuracy of logistic regression classifier: ", logreg1.score(X_test,y_test))
```

```python
logreg2 = LogisticRegression(max_iter = 60).fit(X_train,y_train)
y_pred2 = logreg2.predict(X_test)
print("Accuracy of logistic regression classifier: ", logreg2.score(X_test,y_test))
```

```python
confusion_matrix(y_pred1,y_test)
```

```python
confusion_matrix(y_pred2,y_test)
```

```python
tol = [0.01, 0.001, 0.0001]
max_iter = [100,150,200]
param_grid = dict(tol=tol, max_iter=max_iter)
grid_model = GridSearchCV(estimator=logreg1, param_grid=param_grid, cv=5)
grid_model_result = grid_model.fit(X, y)
best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))
```
